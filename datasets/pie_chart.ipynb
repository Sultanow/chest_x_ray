```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

matplotlib.rcParams.update({
        # "pgf.texsystem": "pdflatex",
        # 'pgf.rcfonts': False,
        'font.family': 'default',
        'text.usetex': True,
    })

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

recipe = ["MIMIC-CXR",
          "CheXpert",
          "ChestX-ray14",
          "PadChest",
          "Open-I",
          "Kermany"]

data = [225, 90, 50, 60, 100, 5]

wedges, texts = ax.pie(data, startangle=-40) # wedgeprops=dict(width=0.4) -> Donut

bbox_props = dict(boxstyle="square,pad=0.4", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")

for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(recipe[i], xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)

ax.set_title("Zusammensetzung des Datensatzes")

# plt.savefig('/mnt/c/Users/Jan/Daten/Dropbox/Master/3_Semester/Masterarbeit/Latex/python_output/kuchendiagramm_datensätze.pdf')

# plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams.update({
        "pgf.texsystem": "pdflatex",
        'pgf.rcfonts': False,
        'font.family': 'serif',
        'text.usetex': True,
    })

# explanation for autopct callback
# https://stackoverflow.com/questions/6170246/how-do-i-use-matplotlib-autopct
def make_autopct(values):
    def my_autopct(pct):
        # explanation for missing % when test.usetex = True
        # https://github.com/matplotlib/matplotlib/issues/4307/
        return r'{:1.1f}\%'.format(pct)
    return my_autopct


# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = "MIMIC-CXR", "CheXpert", "ChestX-ray14", "PadChest", "Open-I", "Kermany"
sizes = [33614, 9340, 2862, 10440, 1387, 5857]
colors = ['#66b3ff','#dbc3b6','#99ff99','#ffcc99', '#c2c2c2','#ff9999']

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors, autopct=make_autopct(sizes), startangle=50, wedgeprops=dict(width=0.35, edgecolor="w"), pctdistance=0.82)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.text(0,0,"63.500 Bilder", ha='center', va='center', fontsize='x-large')

# ax.set_title("Zusammensetzung des Datensatzes")

# plt.show()

plt.savefig('/mnt/c/Users/Jan/Daten/Dropbox/Master/3_Semester/Masterarbeit/Latex/python_output/kuchendiagramm_datensätze.pgf')
```
